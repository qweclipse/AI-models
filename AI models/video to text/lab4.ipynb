```python
import os
import urllib.request
import tarfile
from bs4 import BeautifulSoup
import lxml.html as html
```

```python
# URL сайта с данными
voxforge_url = 'http://www.repository.voxforge1.org/downloads/Russian/Trunk/Audio/Main/16kHz_16bit/'
target_folder = r'D:\Рабочий стол\jupyter\lab4\Voxforge'  # Папка для сохранения данных

# Функция для создания папки, если её нет
def create_folder(folder_path):
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
        print(f"Создана папка: {folder_path}")
    else:
        print(f"Папка уже существует: {folder_path}")

# Функция для загрузки и распаковки архива
def download_and_extract(url, target_folder):
    filename = url.split('/')[-1]  # Имя файла из URL
    filepath = os.path.join(target_folder, filename)
    
    # Загрузка файла
    print(f"Загрузка файла: {url}")
    urllib.request.urlretrieve(url, filepath)
    print(f"Файл загружен: {filepath}")
    
    # Распаковка файла
    if filepath.endswith('.tar.gz') or filepath.endswith('.tgz'):
        print(f"Распаковка файла: {filepath}")
        with tarfile.open(filepath, 'r:gz') as tar:
            tar.extractall(path=target_folder)
        print(f"Файл распакован в: {target_folder}")
        os.remove(filepath)  # Удаляем архив после распаковки
        print(f"Архив удалён: {filepath}")
    else:
        print(f"Файл не является архивом: {filepath}")

# Парсинг страницы и получение ссылок
def get_file_links(base_url):
    print(f"Получение списка файлов с {base_url}")
    response = urllib.request.urlopen(base_url)
    html = response.read().decode('utf-8')
    soup = BeautifulSoup(html, 'html.parser')
    
    # Ищем все ссылки на .tgz файлы
    links = [base_url + a['href'] for a in soup.find_all('a', href=True) if a['href'].endswith('.tgz')]
    print(f"Найдено {len(links)} архивов для загрузки.")
    return links

# Основной процесс
def download_voxforge_dataset(base_url, target_folder):
    create_folder(target_folder)  # Создаём папку для данных
    links = get_file_links(base_url)  # Получаем список ссылок
    
    # Загрузка и распаковка каждого архива
    for link in links:
        download_and_extract(link, target_folder)

# Запуск загрузки
download_voxforge_dataset(voxforge_url, target_folder)
```

```python
import os
import csv

# Укажите путь к основной папке, где находятся ваши данные
base_folder = r"D:\Рабочий стол\jupyter\lab4\Voxforge"

# Итоговый CSV файл
output_csv = os.path.join(base_folder, "dataset.csv")

# Функция для обработки одной папки
def process_folder(folder_path):
    wav_folder = os.path.join(folder_path, "wav")  # Путь к аудиофайлам
    etc_folder = os.path.join(folder_path, "etc")  # Путь к текстовым данным
    prompts_file = os.path.join(etc_folder, "prompts-original")  # Файл с текстами

    # Проверяем наличие необходимых папок и файлов
    if not os.path.exists(wav_folder) or not os.path.exists(prompts_file):
        return []

    # Читаем расшифровки из файла prompts-original
    transcripts = {}
    try:
        with open(prompts_file, "r", encoding="utf-8") as file:
            for line in file:
                parts = line.strip().split(" ", 1)
                if len(parts) == 2:
                    file_name, transcription = parts
                    transcripts[file_name] = transcription
        print(f"Файл '{prompts_file}' успешно прочитан.")
    except Exception as e:
        print(f"Ошибка при чтении файла '{prompts_file}': {e}")
        return []

    # Создаем список записей для CSV
    dataset = []
    for wav_file in os.listdir(wav_folder):
        if wav_file.endswith(".wav"):
            file_id = os.path.splitext(wav_file)[0]  # Убираем расширение
            if file_id in transcripts:
                wav_path = os.path.join(wav_folder, wav_file)
                transcription = transcripts[file_id]
                dataset.append((wav_path, transcription))

    return dataset

# Основной процесс обработки всех папок
all_data = []
for root, dirs, files in os.walk(base_folder):
    if "wav" in dirs and "etc" in dirs:  # Если в папке есть subdirs wav и etc
        all_data.extend(process_folder(root))

# Сохраняем результаты в CSV файл с явным указанием кодировки utf-8
with open(output_csv, mode="w", encoding="utf-8-sig", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["audio_path", "transcription"])  # Заголовки
    writer.writerows(all_data)

print(f"Данные успешно сохранены в {output_csv}")
```

```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split

# Укажите путь к основной папке, где находятся данные
base_folder = r"D:\Рабочий стол\jupyter\lab4\Voxforge"
input_csv = os.path.join(base_folder, "dataset.csv")  # Исходный файл с данными
csv_folder = base_folder  # Папка для сохранения train, val, test

# Проверка существования исходного файла
if not os.path.exists(input_csv):
    raise FileNotFoundError(f"Исходный файл {input_csv} не найден. Убедитесь, что он существует.")

# Загрузка исходных данных
df = pd.read_csv(input_csv, encoding="utf-8")

# Разделение данных на тренировочные, валидационные и тестовые наборы
train, temp = train_test_split(df, test_size=0.2, random_state=42)  # 80% на обучение
val, test = train_test_split(temp, test_size=0.5, random_state=42)  # 10% на валидацию, 10% на тест

# Путь для сохранения файлов
train_csv = os.path.join(csv_folder, "train.csv")
val_csv = os.path.join(csv_folder, "val.csv")
test_csv = os.path.join(csv_folder, "test.csv")

# Сохранение данных с кодировкой utf-8-sig
train.to_csv(train_csv, index=False, encoding="utf-8-sig")
val.to_csv(val_csv, index=False, encoding="utf-8-sig")
test.to_csv(test_csv, index=False, encoding="utf-8-sig")

print(f"Файлы успешно созданы:")
print(f"Тренировочный набор: {train_csv}")
print(f"Валидационный набор: {val_csv}")
print(f"Тестовый набор: {test_csv}")
```

```python
import os
from yt_dlp import YoutubeDL
from transformers import Wav2Vec2Processor, Wav2Vec2ForCTC
import torch
import subprocess

# Настройка устройства
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
MODEL_NAME = "facebook/wav2vec2-base-960h"

processor = Wav2Vec2Processor.from_pretrained(MODEL_NAME)
model = Wav2Vec2ForCTC.from_pretrained(MODEL_NAME).to(DEVICE)

print(f"Используемое устройство: {DEVICE}")

# Укажите папку, где будет сохраняться аудио
OUTPUT_FOLDER = r"D:\Рабочий стол\jupyter\audio"

def download_audio(youtube_url, output_folder=OUTPUT_FOLDER):
    """
    Скачивает аудио из YouTube-видео.
    """
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # Путь для временного файла
    temp_audio_path = os.path.join(output_folder, "temp_audio.webm")
    wav_path = os.path.join(output_folder, "audio.wav")
    
    # Удаляем старые временные файлы, если они существуют
    if os.path.exists(temp_audio_path):
        os.remove(temp_audio_path)
    if os.path.exists(wav_path):
        os.remove(wav_path)
    
    options = {
        "format": "bestaudio/best",
        "outtmpl": temp_audio_path,
        "quiet": False,
    }
    with YoutubeDL(options) as ydl:
        ydl.download([youtube_url])
    
    # Проверяем, существует ли скачанный файл
    if not os.path.exists(temp_audio_path):
        raise FileNotFoundError(f"Файл {temp_audio_path} не найден.")
    
    return temp_audio_path

def convert_audio_to_wav_with_ffmpeg(input_audio_path, output_audio_path):
    """
    Конвертирует аудио в формат WAV с помощью FFmpeg.
    """
    try:
        command = [
            "ffmpeg", "-i", input_audio_path,  # Исходный файл
            "-ar", "16000",  # Частота дискретизации
            "-ac", "1",      # Один канал (моно)
            output_audio_path,  # Выходной файл
            "-y"             # Перезапись файла
        ]
        subprocess.run(command, check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    except subprocess.CalledProcessError as e:
        raise RuntimeError(f"Ошибка конвертации аудио: {e.stderr.decode()}")

def transcribe_audio(audio_path):
    """
    Выполняет расшифровку текста из аудиофайла.
    """
    try:
        import torchaudio  # Импортируем здесь, чтобы использовать только для расшифровки
        waveform, sample_rate = torchaudio.load(audio_path)
        resampler = torchaudio.transforms.Resample(orig_freq=sample_rate, new_freq=16000)
        waveform = resampler(waveform).squeeze(0).to(DEVICE)
        input_values = processor(waveform, sampling_rate=16000, return_tensors="pt").input_values.to(DEVICE)
        logits = model(input_values).logits
        predicted_ids = torch.argmax(logits, dim=-1)
        transcription = processor.batch_decode(predicted_ids)[0]
        return transcription
    except Exception as e:
        raise RuntimeError(f"Ошибка при расшифровке аудио {audio_path}: {e}")

def youtube_to_text(youtube_url):
    """
    Скачивает видео, конвертирует аудио в WAV и выполняет расшифровку текста.
    """
    print("[INFO] Начинаем обработку...")
    try:
        temp_audio_path = download_audio(youtube_url)
        wav_audio_path = os.path.join(OUTPUT_FOLDER, "audio.wav")
        convert_audio_to_wav_with_ffmpeg(temp_audio_path, wav_audio_path)
        transcription = transcribe_audio(wav_audio_path)
        print("[INFO] Расшифровка завершена.")
        return transcription
    except Exception as e:
        print(f"[ERROR] Произошла ошибка: {e}")
        return None

if __name__ == "__main__":
    youtube_url = input("Введите ссылку на YouTube-видео: ")
    transcription = youtube_to_text(youtube_url)

    if transcription:
        print("\n[Текст из видео]:\n")
        print(transcription)
    else:
        print("[ERROR] Расшифровка не выполнена.")
```
